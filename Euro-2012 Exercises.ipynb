```python
def my_check(euro12):
    if euro12["Goals"] <= 2:
        return "Below Avg"
    elif ((euro12["Goals"] > 2) and (euro12["Goals"] <= 5)):
        return "Average"
    elif ((euro12["Goals"] > 5) and (euro12["Goals"] <= 10)):
        return "Above Average"
    else:
        return "Excellent"
euro12["Performance"] = euro12.apply(lambda goal: my_check(goal), axis = 1)
euro12
```
